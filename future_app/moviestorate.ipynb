```python
def moviestorate():
    url = "http://www.omdbapi.com/?i="
    movie_data = []
    for index, row in popular.df.iterrows():
        imdb = row["imdbId"]
        response = requests.get(url + str(imdb) + config.api_key)
        data = response.json()
        if "Error" not in data.keys() and data["Poster"] != "N/A":
            movie_data.append(data)
            print(len(movie_data))
    return(movie_data)
```
